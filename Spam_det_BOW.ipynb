```python
import pandas as pd
```

```python
import numpy as np
```

```python
mails = pd.read_csv('spamails.csv')
mails = mails.drop(['Unnamed: 0','label_num'],axis=1)
```

```python
mails
```

```python
mails.columns
```

```python
mails.rename(columns = {'label':'Category','text':'Message'}, inplace=True)
```

```python
mails
```

```python
mails['Category'].value_counts()
```

```python
mails['spam']= mails['Category'].apply(lambda x : 1 if x=='spam' else 0)
```

```python
mails
```

```python
mails['Message']= mails['Message'].apply(lambda x : x[9:])
mails
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(mails.Message, mails.spam, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
```

```python
X_traincv = v.fit_transform(X_train.values)
X_traincv
```

```python
X_traincv.shape
```

```python
v.get_feature_names_out()[9000:9050]
```

```python
v.vocabulary_
```

```python
v.get_feature_names_out()[11244]
```

```python
X_traincv = X_traincv.toarray()
```

```python
X_traincv[0]
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_traincv, y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```

```python
X_test_cv_1 = X_test_cv.toarray()
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv_1)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
```
