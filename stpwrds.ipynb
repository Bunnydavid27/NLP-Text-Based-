```python
## PREPROCESS
stop_words = STOP_WORDS
```

```python
stop_words
```

```python
stp = list(stop_words)
```

```python
stp
```

```python
def preprocess(text):
    nlp = spcy(text)
    cleaned_text = [token.text for token in nlp if not token.is_stop] 
    return ' '.join(cleaned_text)
```

```python
preprocess('We just opened our wings')
```
